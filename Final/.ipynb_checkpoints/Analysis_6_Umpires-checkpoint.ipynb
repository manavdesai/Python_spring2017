# Analysis on Umpires

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
path=os.getcwd()
path
```

```python
path=path+'\\Original_Data\\'
path
```

```python
df_matches= pd.read_csv(path + "matches.csv")
```

```python
umpire=pd.concat([df_matches['umpire1'],df_matches['umpire2']]) 
```

```python
fig, top = plt.subplots(figsize=(10,7))

top=umpire.value_counts().head(7).plot.bar(width=0.4,color='brown')


top.set_xlabel("Umpires", size=20)
top.set_title("Top Umpires", size=20)

for each in top.patches:
    top.annotate(format(each.get_height()), (each.get_x()+0.15, each.get_height()+0.25))
plt.show()
```

```python
extras=df_matches[['umpire1','umpire2']].sum()
sizes=[5161,680,]
plt.pie(sizes, labels=['Umpire 1','Umpire 2'],
        colors=['yellow', 'green'],explode=(0,0,),autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Percentage of Extras")
fig = plt.gcf()
fig.set_size_inches(8,8)
plt.plot()
plt.show()
```
