```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
path=os.getcwd()
path
```

```python
path=path+'\\Original_Data\\'
path
```

```python
df_matches= pd.read_csv(path + "matches.csv")
df_deliveries= pd.read_csv(path + "deliveries.csv")
```

```python
fig, top = plt.subplots(figsize=(12,7))
top = df_matches['venue'].value_counts().plot.bar(width=.8, color=["blue", "gold", "red", "green"])
top.set_xlabel('Name of Stadiums',size=20)
top.set_ylabel('Count', size=20)
plt.show()
```
