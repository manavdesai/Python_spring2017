## 1. Top Run scorers

```python
runs_df = df_deliveries.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
runs_df = runs_df.iloc[:10,:]
```

### Conclusion
##### ~ V Kohli is the top run scorer amongst all followed by SK Raina.
##### ~ CH Gayle and DA Warner are highest run scorers in foreign players.

## 2. Performance of Top 5 Batman over the seasons

```python
top=df_batsman.groupby(['season','batsman'])['batsman_runs'].sum().reset_index()
top=top.groupby(['season','batsman'])['batsman_runs'].sum().unstack().T
top['Total']=top.sum(axis=1)
top=top.sort_values(by='Total',ascending=0)[:5]
top.drop('Total',axis=1,inplace=True)
top.T.plot(figsize=(12,7))
plt.show()
```

### Conclusion
##### ~ V Kohli shows a considerable improved performance in the year 2016
##### ~ CH Gayle had good performance from 2011 to 2013 but then showed a drop
##### ~ SK Raina and RG Sharma had nearly consistent performance all the seasons

## 3. Players with maximum number of 4's

```python
#Creating a data frame having players and the numbers of Fours by them
fours_df = df_deliveries.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
fours_df = fours_df.iloc[:10,:]
fours_df.head()
```

```python
batsman_array = np.array(fours_df['batsman'])
length = np.arange(len(batsman_array))
fig, top = plt.subplots(figsize=(12,7))
bars = top.bar(length, np.array(fours_df['batsman_runs']), width=0.7, color= "Maroon")
top.set_xticks(length+(0.7/2.))
top.set_xticklabels(batsman_array, rotation='vertical')
top.set_xlabel("Players",size=20)
top.set_ylabel("No of fours", size= 20)
top.set_title("Players with maximum number of Fours!", size=20)
autolabel(bars)
plt.show()
```

### Conclusion 
##### ~  G Gambhir has hit maximum number of FOURS. Total of 422 Fours

## 4. Players with maximum number of 6's

```python
#Creating a data frame having players and the numbers of Sixes by them
six_df = df_deliveries.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
six_df = six_df.iloc[:10,:]
six_df.head()
```

```python
#Plotting the bar graph to represent the Top 10 players with highest number of sixes 
batsman_array = np.array(six_df['batsman'])
length = np.arange(len(batsman_array))
fig, top = plt.subplots(figsize=(12,7))
bars = top.bar(length, np.array(six_df['batsman_runs']), width=0.7, color= "Purple")
top.set_xticks(length+(0.7/2.))
top.set_xticklabels(batsman_array, rotation='vertical')
top.set_xlabel("Players",size=20)
top.set_ylabel("No of Sixes", size= 20)
top.set_title("Players with maximum number of Sixes!", size=20)
autolabel(bars)
plt.show()
```

### Conclusion 
##### ~ CH Gayle has hit maximum number of SIXES. Total 252 Sixes.

## 5. Players with maximum number of Dot balls

```python
#Creating a data frame having players with maximum dot balls
dot_df = df_deliveries.groupby('batsman')['batsman_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='batsman_runs', ascending=False).reset_index(drop=True)
dot_df = dot_df.iloc[:10,:]
dot_df.head()
```

```python
#Plotting the bar graph to represent the Top 10 players with highest number of sixes 
batsman_array = np.array(dot_df['batsman'])
length = np.arange(len(batsman_array))
fig, top = plt.subplots(figsize=(10,7))
bars = top.bar(length, np.array(dot_df['batsman_runs']), width=0.7, color= "red")
top.set_xticks(length+(0.7/2.))
top.set_xticklabels(batsman_array, rotation='vertical')
top.set_xlabel("Players",size=20)
top.set_ylabel("No of Dot balls", size= 20)
top.set_title("Players with maximum number of Dot balls!", size=20)
autolabel(bars)
plt.show()
```

### Conclusion
##### ~ Surprisingly, V Kohli, who has scored the maximum runs is also the player with maximum DOT balls
